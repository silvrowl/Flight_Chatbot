```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
def url_to_string(url):
    res = requests.get(url)
    html = res.text
    soup = BeautifulSoup(html, 'html5lib')
    for script in soup(["script", "style", 'aside']):
        script.extract()
    return " ".join(re.split(r'[\n\t]+', soup.get_text()))
```

```python
with open('Rain_book.txt', 'r',encoding='utf-8') as myfile:
    rain_data = myfile.read()
```

```python
rain_data[0:100]
```

```python
rainbook = rain_data.split('APPENDIX')[0].strip()
```

```python
rainbook_data = re.sub('(\t)', ' ', rainbook)# remove the tabs
```

```python
para = [re.sub('(\n)',' ',t) for t in rainbook_data.split('.\n')] #split into paragraphs and then remove the \n
```

```python
para = [re.sub(u"(\ufeff|\xa0)", " ", t) for t in para] # remove some of the unicode controls
```

```python
sentences = [nltk.sent_tokenize(sent) for sent in para]
```

```python
len(sentences)
```

```python
stop_words = stopwords.words('english')
```

```python
stop_words = stop_words + ['com', 'edu', 'subject', 'lines', 'organization', 'would', 'article', 'could','in','of']
```

```python
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))  # deacc=True removes punctuations
```

```python
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in texts]
```

```python
# Build the bigram and trigram models
bigram = gensim.models.Phrases(para, min_count=5, threshold=100) # higher threshold fewer phrases.
trigram = gensim.models.Phrases(bigram[para], threshold=100)  

# Faster way to get a sentence clubbed as a trigram/bigram
bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)
```

```python
def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]
```

```python
def lemmatization(texts, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    """https://spacy.io/api/annotation"""
    texts_out = []
    for sent in texts:
        doc = nlp(" ".join(sent)) 
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
    return texts_out
```

```python
nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])
```

```python
texts = list(sent_to_words(para))
```

```python
texts_nostops = remove_stopwords(texts)
```

```python
print(texts[0:1])
print(texts_nostops[0:1])
```

```python
# Form Bigrams
texts_bigrams = make_bigrams(texts_nostops)
```

```python
texts_bigrams[0:1]
```

```python
texts_lemmatized = lemmatization(texts_bigrams, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV'])
```

```python
# Create dictionary; expects array as input
dct = corpora.Dictionary(texts_lemmatized)
```

```python
dct.filter_extremes(no_below=10, no_above=0.2)
```

```python
# Get information about the dictionary
print(dct)
```

```python
# Show the word to id map
print(dct.token2id)
```

```python
mycorpus = [dct.doc2bow(doc, allow_update=True) for doc in texts_lemmatized]
```

```python
word_counts = [[(dct[id], count) for id, count in line] for line in mycorpus]
```

```python
word_counts[0:10]
```

###LDA

```python
lda_model = LdaMulticore(corpus=mycorpus,
                         id2word=dct,
                         random_state=100,
                         num_topics=4,
                         passes=10,
                         chunksize=1000,
                         batch=False,
                         alpha='asymmetric',
                         decay=0.5,
                         offset=64,
                         eta=None,
                         eval_every=0,
                         iterations=100,
                         gamma_threshold=0.001,
                         per_word_topics=True)
```

```python
# save the model
lda_model.save('lda_model.model')

# See the topics
lda_model.print_topics(-1)
```

```python
for c in lda_model[mycorpus[5:9]]:
    print("Document Topics      : ", c[0])      # [(Topics, Perc Contrib)]
    print("Word id, Topics      : ", c[1][:3])  # [(Word id, [Topics])]
    print("Phi Values (word id) : ", c[2][:2])  # [(Word id, [(Topic, Phi Value)])]
    print("Word, Topics         : ", [(dct[wd], topic) for wd, topic in c[1][:2]])   # [(Word, [Topics])]
    print("Phi Values (word)    : ", [(dct[wd], topic) for wd, topic in c[2][:2]])  # [(Word, [(Topic, Phi Value)])]
    print("------------------------------------------------------\n")
```

```python
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lda_model, mycorpus, dct)
vis
```
